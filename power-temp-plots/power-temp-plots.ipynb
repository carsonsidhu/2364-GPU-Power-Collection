# Resnet

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a DataFrame
df = pd.read_csv('./gpu_metrics_resnet50_v2.csv', header=None)
df.columns = ['Power (W)', 'Temperature (°C)']
df['Time (s)'] = df.index * 0.5
```

```python
start_time = 0  # seconds
end_time = 700  # seconds
start_row = int(start_time / 0.5)
end_row = int(end_time / 0.5)

# Ensure we don't exceed the DataFrame's length
if end_row > len(df):
    end_row = len(df)
    if start_row >= end_row:
        start_row = end_row - 1  # Ensure at least one row is included

average_power_test = df['Power (W)'].iloc[start_row:end_row].mean()
average_temp_test = df['Temperature (°C)'].iloc[start_row:end_row].mean()

energy_window_test = np.trapz(df['Power (W)'].iloc[start_row:end_row], dx=0.5)



start_time = 701  # seconds
end_time = 1043  # seconds
start_row = int(start_time / 0.5)
end_row = int(end_time / 0.5)

# Ensure we don't exceed the DataFrame's length
if end_row > len(df):
    end_row = len(df)
    if start_row >= end_row:
        start_row = end_row - 1  # Ensure at least one row is included

average_power_idle = df['Power (W)'].iloc[start_row:end_row].mean()
average_temp_idle = df['Temperature (°C)'].iloc[start_row:end_row].mean()
```

```python
# Creating separate plots for Power and Temperature

# Plot for Power
plt.figure(figsize=(14, 6))
plt.plot(df['Time (s)'], df['Power (W)'], label='Power (W)')
plt.axvline(x=0, linestyle='--', color='red', label='Start Time')  # Start time line
plt.axvline(x=700, linestyle='--', color='green', label='End Time')  # End time line 
plt.axhline(y=average_power_test, linestyle='--', color='blue', label=f'Average Test Power')
plt.axhline(y=average_power_idle, linestyle='--', color='orange', label=f'Average Idle Power')
plt.title('Power vs Time (Resnet)')
plt.xlabel('Time (s)')
plt.ylabel('Power (W)')
plt.grid(True)
plt.legend()

# Show the first plot
plt.show()

# Total energy expended (for the entire time period)
total_energy = np.trapz(df['Power (W)'], dx=0.5)  # dx is the time step in seconds

print(f"Total Energy Used: {total_energy}J")
print(f"Energy Used During Test: {energy_window_test}J")

frames = 73
energy_per_frame = energy_window_test/frames

print(f"Energy per frame: {energy_per_frame}J/frame")
print(f"Speed: {frames/700} frames per second")
```

```python
# Plot for Temperature
plt.figure(figsize=(14, 6))
plt.plot(df['Time (s)'], df['Temperature (°C)'], label='Temperature (°C)', color='red')
plt.axvline(x=0, linestyle='--', color='gray', label='Start Time')  # Start time line
plt.axvline(x=700, linestyle='--', color='green', label='End Time')  # End time line 
plt.axhline(y=average_temp_test, linestyle='--', color='purple', label=f'Average Test Temperature')
plt.axhline(y=average_temp_idle, linestyle='--', color='red', label=f'Average Idle Temperature')
plt.title('Temperature vs Time (Resnet)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()

# Show the second plot
plt.show()
```
